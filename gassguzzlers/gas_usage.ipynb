```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('avg_gas_used.csv')
```

```python
df.head()
```

```python
plt.figure(figsize = (30,10))

x=df['Date']
y=df['Gas usage']
plt.xlabel('Date in Month and Year')
plt.ylabel('Average usage')
plt.xticks(rotation=90)

plt.plot(x,y)
plt.show()
```
