```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('num_transactions.csv')
df = df.sort_values(by='Date')
```

```python
df.head() 
```

```python
plt.figure(figsize = (30,10))
plt.bar(df['Date'],df['Number of transactions'])
plt.xlabel('Date in Month and Year')
plt.ylabel('Total Number of Transactions each month')
plt.xticks(rotation=90)
plt.figure(figsize = (50,5))
```
